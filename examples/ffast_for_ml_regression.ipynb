# Data

```python
from pandas import read_csv
data = read_csv("/kaggle/input/contexttopic-classification-from-statement/Context.csv")
data.head()
```

# Preprocessing (with ffast)

```python
!pip install ffast
```

```python
from ffast import load
tokeniser = load('poincare')
```

```python
small_data = data.head(100)
small_data['input'] = small_data['Text'].map(lambda text:tokeniser.encode(text).vector)
small_data['output'] = small_data['Context/Topic'].map(lambda topic:tokeniser.encode(topic).semantics()[-1])
small_data.head()
```

# Training (ML)

```python
from sklearn.neural_network import MLPRegressor
classifier = MLPRegressor(verbose=True, solver='adam', activation='tanh', tol=1e-8, n_iter_no_change=50, hidden_layer_sizes=(1000)*3)
classifier.fit(small_data['input'].to_list(),small_data['output'].to_list())
```

```python
from matplotlib.pyplot import plot
plot(classifier.loss_curve_)
```

# Topic Classification

```python
topic_classification = lambda text: str(tokeniser.decode_semantics(classifier.predict([tokeniser.encode(text).vector])))
```

```python
topic_classification("i love to listen to a good song")
```
